#### Quelle est la diffrence entre append() et extend() en Python ?
Les deux mthodes permettent d'ajouter des lments dans une liste, mais ne fonctionnent pas de la mme faon.

Le langage Python fournit plusieurs mthodes pour grer les listes. append() et extend() en font partie. Elles permettent d'ajouter des lments dans une liste, mais ne fonctionnent pas de la mme faon. Elles ont un rsultat diffrent sur la liste.

La mthode append() ajoute les lments dans une liste mais conserve la forme d'itrable. L'itrable sera ajout comme un nouvel lment de la liste. Il y aura donc une liste  2 dimensions.

```python
maListe = [1, 2, 3]
maListe.append([4, 5])
print(maListe)
```

Pour accder  la valeur 4, il faudra crire maListe[3][0];

La mthode extend() extrait les lments de l'itrable transmis en argument, agrandit la taille de la liste pour qu'elle puisse contenir les lments extraits, puis les ajoute  la suite dans la liste. Pour des objets itrables, son fonctionnement est similaire  l'oprateur +=. Cependant la mthode extend() peut tre appele  la suite d'une autre fonction ou mthode, ce qui n'est pas possible avec l'oprateur +=.

```python
maListe = [1, 2, 3]
maListe.extend([4, 5])
print(maListe)
```

Pour accder  la valeur 4, il faudra crire maListe[3];
